```python
#Load the fasta file
def parse_fasta(fasta_file):
    """Parse a fasta file and return a dictionary."""
    with open(fasta_file) as f:
        lines = f.readlines()
    fasta_dict = {}
    for line in lines:
        if line.startswith(">"):
            key = (line.strip().split()[0][1:]).replace("|","_")
            fasta_dict[key] = ""
        else:
            fasta_dict[key] += line.strip().upper()
    return fasta_dict

msa = parse_fasta("danio_rerio_opsins.aln-fasta")
```
